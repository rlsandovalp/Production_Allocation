```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Mix_plot = 'M3'
Methods = ['original_12', 'original_all', 'newOF_12', 'newOF_all']              # {newOF,direct,original}

data_to_plot = []

for i, Method in enumerate(Methods):
    data_to_plot.append([])
    errors_1CM = pd.read_csv('../Results/Comparison_methods/Nouvelle/'+Method+'/Errors_1CM_rel.txt', header = None)
    errors_2CM = pd.read_csv('../Results/Comparison_methods/Nouvelle/'+Method+'/Errors_2CM_rel.txt', header = None)
    errors_3CM = pd.read_csv('../Results/Comparison_methods/Nouvelle/'+Method+'/Errors_3CM_rel.txt', header = None)
    errors_4CM = pd.read_csv('../Results/Comparison_methods/Nouvelle/'+Method+'/Errors_4CM_rel.txt', header = None)
    errors_5CM = pd.read_csv('../Results/Comparison_methods/Nouvelle/'+Method+'/Errors_5CM_rel.txt', header = None)
    errors_6CM = pd.read_csv('../Results/Comparison_methods/Nouvelle/'+Method+'/Errors_6CM_rel.txt', header = None)
    errors_7CM = pd.read_csv('../Results/Comparison_methods/Nouvelle/'+Method+'/Errors_7CM_rel.txt', header = None)
    errors_8CM = pd.read_csv('../Results/Comparison_methods/Nouvelle/'+Method+'/Errors_8CM_rel.txt', header = None)
    errors_9CM = pd.read_csv('../Results/Comparison_methods/Nouvelle/'+Method+'/Errors_9CM_rel.txt', header = None)

    estimates_1CM = pd.read_csv('../Results/Comparison_methods/Nouvelle/'+Method+'/Estimates_1CM.txt', header = None)
    estimates_2CM = pd.read_csv('../Results/Comparison_methods/Nouvelle/'+Method+'/Estimates_2CM.txt', header = None)
    estimates_3CM = pd.read_csv('../Results/Comparison_methods/Nouvelle/'+Method+'/Estimates_3CM.txt', header = None)
    estimates_4CM = pd.read_csv('../Results/Comparison_methods/Nouvelle/'+Method+'/Estimates_4CM.txt', header = None)
    estimates_5CM = pd.read_csv('../Results/Comparison_methods/Nouvelle/'+Method+'/Estimates_5CM.txt', header = None)
    estimates_6CM = pd.read_csv('../Results/Comparison_methods/Nouvelle/'+Method+'/Estimates_6CM.txt', header = None)
    estimates_7CM = pd.read_csv('../Results/Comparison_methods/Nouvelle/'+Method+'/Estimates_7CM.txt', header = None)
    estimates_8CM = pd.read_csv('../Results/Comparison_methods/Nouvelle/'+Method+'/Estimates_8CM.txt', header = None)
    estimates_9CM = pd.read_csv('../Results/Comparison_methods/Nouvelle/'+Method+'/Estimates_9CM.txt', header = None)

    estimates_1CM['errors'] = errors_1CM
    estimates_1CM.set_index(0, inplace = True)
    data1 = estimates_1CM.loc[Mix_plot].values[:,3]

    estimates_2CM['errors'] = errors_2CM
    estimates_2CM.set_index(0, inplace = True)
    data2 = estimates_2CM.loc[Mix_plot].values[:,3]

    estimates_3CM['errors'] = errors_3CM
    estimates_3CM.set_index(0, inplace = True)
    data3 = estimates_3CM.loc[Mix_plot].values[:,3]

    estimates_4CM['errors'] = errors_4CM
    estimates_4CM.set_index(0, inplace = True)
    data4 = estimates_4CM.loc[Mix_plot].values[:,3]

    estimates_5CM['errors'] = errors_5CM
    estimates_5CM.set_index(0, inplace = True)
    data5 = estimates_5CM.loc[Mix_plot].values[:,3]

    estimates_6CM['errors'] = errors_6CM
    estimates_6CM.set_index(0, inplace = True)
    data6 = estimates_6CM.loc[Mix_plot].values[:,3]

    estimates_7CM['errors'] = errors_7CM
    estimates_7CM.set_index(0, inplace = True)
    data7 = estimates_7CM.loc[Mix_plot].values[:,3]

    estimates_8CM['errors'] = errors_8CM
    estimates_8CM.set_index(0, inplace = True)
    data8 = estimates_8CM.loc[Mix_plot].values[:,3]

    estimates_9CM['errors'] = errors_9CM
    estimates_9CM.set_index(0, inplace = True)
    data9 = estimates_9CM.loc[Mix_plot].values[3]

    data_to_plot[i] = np.array((data1, data2, data3, data4, data5, data6, data7, data8, data9), dtype="object")
```

```python
plt.style.use('stylesheet.mpltstyle_boxplot')
fig, ax = plt.subplots(2, 2, figsize = (4.3,3.5))

ax[0,0].set_ylabel(r"$MAPE_{Nv}$ [\%]")
ax[0,0].annotate('(C1)', xy=(0.85, 0.9), xycoords="axes fraction")
ax[0,0].boxplot(data_to_plot[0], sym='.', showmeans = True)
ax[0,0].set_ylim(0,100)

ax[0,1].annotate('(C2)', xy=(0.85, 0.9), xycoords="axes fraction")
ax[0,1].boxplot(data_to_plot[1], sym='.', showmeans = True)
ax[0,1].set_ylim(0,100)

ax[1,0].set_ylabel(r"$MAPE_{Nv}$ [\%]")
ax[1,0].set_xlabel('Number of Synthetic Mixtures, ' r'$N_{SM}$')
ax[1,0].annotate('(C3)', xy=(0.85, 0.9), xycoords="axes fraction")
ax[1,0].boxplot(data_to_plot[2], sym='.', showmeans = True)
ax[1,0].set_ylim(0,100)


ax[1,1].set_xlabel('Number of Synthetic Mixtures, ' r'$N_{SM}$')
ax[1,1].annotate('(C4)', xy=(0.85, 0.9), xycoords="axes fraction")
ax[1,1].boxplot(data_to_plot[3], sym='.', showmeans = True)
ax[1,1].set_ylim(0,100)

plt.savefig('../Figures/Boxplots_nouvelle.pdf')
```
