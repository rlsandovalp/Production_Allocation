```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
folder = './../Data_Base/Maria/'
file1 = 'Maria_2repetitions'
dataset1 = pd.read_csv(folder+'/'+file1+".csv", sep = ';').drop(columns = 'Replicate').set_index('Mix')
file2 = 'Maria_10repetitions'
dataset2 = pd.read_csv(folder+'/'+file2+".csv", sep = ';').drop(columns = 'Replicate').set_index('Mix')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(12, 2.5))
fluid = 'M6'
ax.scatter(range(110), dataset1.loc[fluid].mean(), s = 3)
ax.scatter(range(110), dataset2.loc[fluid].mean(), s = 3)
# ax.scatter(range(110), dataset1.loc['EM1'].values[0,:], s = 3)
ax.scatter(range(110), dataset1.loc[fluid].mean()/dataset2.loc[fluid].mean(), s = 3)  
ax.semilogy()
```
