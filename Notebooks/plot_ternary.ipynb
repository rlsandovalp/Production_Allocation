```python
import ternary
import pandas as pd
import matplotlib.pyplot as plt
```

### McCaffrey

```python
plt.style.use('stylesheet.mpltstyle_ternary')

figure, ax = plt.subplots(figsize = (4.3, 5))
figure.gca().get_xaxis().set_visible(False)
figure.gca().get_yaxis().set_visible(False)

mixtures = ['M1', 'M2', 'M3', 'M4', 'M5', 'M6', 'M7', 'M8', 'M9']
folder = '../Results/NGS_J19_DRK/McCaffrey/'

labels = ["Estimates, " r'$\mathbf{\hat{x}}_{Mc}$', "True value, " r'$\mathbf{x}^{*}$', "Estimates, " r'$\bar{\bar{\mathbf{x}}}_{Mc}$', "Compositional mean, " r'$\bar{\mathbf{x}}_{Mc}$']

true = pd.read_csv(folder+'../real_results.txt', header = None, sep = ',').set_index(0)
averaged =  pd.read_csv(folder+'Average_before_deconvolution.txt', header = None, sep = ',').set_index(0)
compositional_mean = pd.read_csv(folder+'compositional_mean.txt', header = None, sep = ',').set_index(0)

for m, mixture in enumerate(mixtures):
    deconvolutions = pd.read_csv(folder+mixture+'.txt', sep = ' ').values
    ax = figure.add_subplot(4,3,m+1)
    figure, tax = ternary.figure(ax = ax, scale=100)

    # Draw Boundary and Gridlines
    tax.boundary(linewidth = 0.5)
    tax.gridlines(color="gray", multiple=25)

    # Set Axis labels and Title
    offset = 0.16
    tax.set_title(mixture, loc='left', y=0.7, fontweight="bold")
    tax.left_axis_label("EM3 [%]", offset=offset)
    tax.right_axis_label("EM2 [%]", offset=offset)
    tax.bottom_axis_label("EM1 [%]", offset=offset)
    tax.scatter(deconvolutions*100, marker='.', color = 'gray')
    tax.scatter([true.loc[mixture].values], marker='o', color = 'green')
    tax.scatter([averaged.loc[mixture].values], marker='o', color = 'red')
    tax.scatter([compositional_mean.loc[mixture].values], marker='x', color = 'black')

    # Set ticks
    tax.ticks(axis='lbr', multiple=25, offset=0.03, fontsize = 6, linewidth = 0.5)
    tax.get_axes().axis('off')

mixture = 'M10'
deconvolutions = pd.read_csv(folder+mixture+'.txt', sep = ' ').values
ax = figure.add_subplot(4,3,10)
figure, tax = ternary.figure(ax = ax, scale=100)

# Draw Boundary and Gridlines
tax.boundary(linewidth = 0.5)
tax.gridlines(color="gray", multiple=25)

# Set Axis labels and Title
offset = 0.16
tax.set_title(mixture, loc='left', y=0.7, fontweight="bold")
tax.left_axis_label("EM3 [%]", offset=offset)
tax.right_axis_label("EM2 [%]", offset=offset)
tax.bottom_axis_label("EM1 [%]", offset=offset)
tax.scatter(deconvolutions*100, marker='.', label="Estimates, " r'$\mathbf{\hat{x}}_{Mc}$', color = 'gray')
tax.scatter([true.loc[mixture].values], marker='o', label="True value, " r'$\mathbf{x}^{*}$', color = 'green')
tax.scatter([averaged.loc[mixture].values], marker='o', label="Estimates, " r'$\bar{\bar{\mathbf{x}}}_{Mc}$', color = 'red')
tax.scatter([compositional_mean.loc[mixture].values], marker='x', label="Compositional mean, " r'$\bar{\mathbf{x}}_{Mc}$', color = 'black')

# Set ticks
tax.ticks(axis='lbr', multiple=25, offset=0.03, fontsize = 6, linewidth = 0.5)
tax.get_axes().axis('off')

figure.legend(bbox_to_anchor=[0.8, 0.2])
ternary.plt.savefig('../Figures/McCaffrey_ternary.pdf')
ternary.plt.show()
```
